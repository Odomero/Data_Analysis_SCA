```python
#Question9

GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

GermanCars_dtf= pd.DataFrame(GermanCars)
japaneseCars_dtf= pd.DataFrame(japaneseCars)

cars = pd.concat([GermanCars_dtf,japaneseCars_dtf],keys=('GermanCars','JapaneseCars'))
cars
```

```python
#Question10

Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}

CarPrice_dtf = pd.DataFrame(Car_Price)
carHorsepower_dtf = pd.DataFrame(car_Horsepower)

car_merge = pd.merge(CarPrice_dtf,carHorsepower_dtf, on= 'Company')
car_merge
```

### MATPLOTLIB

```python
path= "../Downloads/company_sales_data.csv"
sales_dtf = pd.read_csv(path)
sales_dtf.sample(5)
```

```python
#Exercise1

import matplotlib.pyplot as plt

x = sales_dtf['month_number']
y = sales_dtf['total_profit']

plt.figure(figsize=(10, 6))
plt.plot(x,y,linewidth=1.0)
plt.xlim(1,12)
plt.title("Total Profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.show()
```

```python
#Exercise2

plt.plot(sales_dtf["facecream"], color='blue', label='Face Cream Sales Data', marker="o")
plt.plot(sales_dtf["facewash"], color='orange', label='Face Wash Sales Data', marker="o")
plt.plot(sales_dtf["toothpaste"], color='green', label='Toothpaste Sales Data', marker="o")
plt.plot(sales_dtf["bathingsoap"], color='red', label='Bathing Soap Sales Data', marker="o")
plt.plot(sales_dtf["shampoo"], color='brown', label='Shampoo Sales Data', marker="o")
plt.plot(sales_dtf["moisturizer"],color='black', label='Moisturizer Sales Data', marker="o")
plt.title("Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales(Unit)")
plt.legend()
plt.show()
```
